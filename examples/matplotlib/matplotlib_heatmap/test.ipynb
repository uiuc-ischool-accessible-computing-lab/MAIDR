```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np

#create map
a = np.random.random((16, 16))

heatmap = plt.imshow(a, cmap='hot', interpolation='nearest')
plt.colorbar(heatmap)

# generate and save heatmap
plt.savefig('heatmap.svg', format='svg', dpi=1200)
plt.show()
```

```python
import json

# Convert the heatmap data to a list
heatmap_data = a.tolist()

# Define the file path
file_path = 'heatmap_data.json'

# Save the heatmap data to the JSON file
with open(file_path, 'w') as file:
    json.dump(heatmap_data, file)
```
