```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(300, 10, 250)
plt.hist(x)

#create and save a svg file
plt.savefig('histogram.svg', format='svg')
plt.show() 
```

```python
import json
# Calculate the histogram bins and counts
counts, bins = np.histogram(x)

# Create a dictionary to store the histogram data
histogram_data = {'counts': counts.tolist(), 'bins': bins.tolist()}

# Save the histogram data as a JSON file
with open('histogram_data.json', 'w') as file:
    json.dump(histogram_data, file)
```
