```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
categories = ['Category 1', 'Category 2', 'Category 3']
data = np.random.randint(1, 10, size=(3, 5))

# Transpose the data array
data = data.T

# Create stacked barplot
fig, ax = plt.subplots()
ax.bar(categories, data[0], label='Data 1')
ax.bar(categories, data[1], bottom=data[0], label='Data 2')
ax.bar(categories, data[2], bottom=data[0]+data[1], label='Data 3')

# Add labels and legend
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Stacked Barplot')
ax.legend()

# Save the plot as SVG file
plt.savefig('stackedbarplot.svg', format='svg')

# Show the plot
plt.show()
```

```python
import json

# Define the data
stacked_barplot_data = {
    'categories': categories,
    'data': data.tolist()
}

# Save the data as JSON
with open('stackedbarplot.json', 'w') as json_file:
    json.dump(stacked_barplot_data, json_file)
```
