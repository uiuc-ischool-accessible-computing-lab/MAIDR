```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)
y = np.random.rand(100)

# Create the line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')

# Display the plot
plt.savefig('lineplot.svg', format='svg')
plt.show()
```

```python
import json

# Create a dictionary to store the line plot data
lineplotdata = {
    'x': x.tolist(),
    'y': y.tolist()
}

# Convert the dictionary to JSON
json_data = json.dumps(lineplotdata)

# Write the JSON data to a file
with open('lineplotdata.json', 'w') as file:
    file.write(json_data)
```
