```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

countries=pd.read_csv('countries.csv')
```

```python
data_2007=countries[countries['year']==2007]
top_10_p=data_2007.sort_values('population',ascending=False).head(10)

plt.figure(figsize=(12,6))

x=range(10)
y=top_10_p['population']/10**9
plt.bar(x,y)
plt.xlabel('Countries')
plt.ylabel('Population in Billions')
plt.title('10 Most Populous Countries in 2007')

plt.gcf()
plt.savefig("barplot.svg", format="svg")
plt.show()
```
